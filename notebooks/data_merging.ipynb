# Data Merging

```python
gear_box_map = {
    'manuell': 'manual',
    'automatik': 'automatic',
}

kaggle_data['gearbox'] = kaggle_data['gearbox'].map(gear_box_map)
kaggle_data['gearbox'].value_counts()
```

```python
crawled_data['gearbox'] = crawled_data['gearbox'].str.lower()
crawled_data['gearbox'].value_counts()
```

```python
# Standardize brand names
kaggle_data['brand'] = kaggle_data['brand'].astype(str).str.lower().str.replace(r'[-_]', '-', regex=True)
kaggle_data['brand'].value_counts().head(50)
```

```python
crawled_data['brand'] = crawled_data['brand'].astype(str).str.lower().str.replace(r'[-_]', '-', regex=True)
crawled_data['brand'].value_counts().head(50)
```

```python
fuel_map_kaggle = {
    'benzin': 'gasoline', 'diesel': 'diesel', 'hybrid': 'hybrid',
    'elektro': 'electric', 'lpg': 'lpg', 'cng': 'cng', 'andere': 'other'
}

kaggle_data['fuelType'] = kaggle_data['fuelType'].str.lower().map(fuel_map_kaggle)
kaggle_data['fuelType'].value_counts()
```

```python
fuel_map_crawled = {
    'Gasoline': 'gasoline', 'Diesel': 'diesel', 
    'Electric/Gasoline': 'hybrid', 'Electric/Diesel': 'hybrid',
    'Electric': 'electric', 'LPG': 'lpg', 'CNG': 'cng'
}
crawled_data['fuelType'] = crawled_data['fuelType'].map(fuel_map_crawled)
crawled_data['fuelType'].value_counts()
```

```python
crawled_data.rename(columns={'mileage': 'kilometer'}, inplace=True)
crawled_data['kilometer'] = pd.to_numeric(crawled_data['kilometer'], errors='coerce').astype('Int64')
print(crawled_data['kilometer'].value_counts(dropna=False))
crawled_data['kilometer'].value_counts()
```

```python
crawled_data['yearOfRegistration'] = crawled_data['year'].astype(str).str.extract(r'(\d{4})').astype("Int64")
crawled_data['yearOfRegistration'].value_counts()
```

```python
vehicleType_map_kaggle = {
    'limousine': 'sedan', 'kleinwagen': 'compact', 'kombi': 'station_wagon',
    'bus': 'van', 'cabrio': 'convertible', 'coupe': 'coupe', 'suv': 'suv', 'andere': 'other'
}
kaggle_data['vehicleType'] = kaggle_data['vehicleType'].str.lower().map(vehicleType_map_kaggle)
kaggle_data['vehicleType'].value_counts()
```

```python
vehicleType_map_crawled = {
    'SUV/Off-Road/Pick-Up': 'suv', 'Sedan': 'sedan', 'Compact': 'compact',
    'Station Wagon': 'station_wagon', 'Van': 'van', 'Convertible': 'convertible',
    'Coupe': 'coupe', 'Other': 'other'
}
crawled_data['vehicleType'] = crawled_data['vehicleType'].map(vehicleType_map_crawled)
crawled_data['vehicleType'].value_counts()
```

```python
seller_map_kaggle = {'privat': 'private', 'gewerblich': 'dealer'}
kaggle_data['seller'] = kaggle_data['seller'].str.lower().map(seller_map_kaggle)
kaggle_data['seller'].value_counts()
```

```python
seller_map_crawled = {'Dealer': 'dealer'}
crawled_data['seller'] = crawled_data['seller'].map(seller_map_crawled)
crawled_data['seller'].value_counts()
```

```python
kaggle_data['model'] = kaggle_data['model'].astype(str).str.lower().str.replace(r'[-_ ]', '-', regex=True)
kaggle_data['model'].value_counts().head(50)
```

```python
crawled_data['model'] = crawled_data['model'].astype(str).str.lower().str.replace(r'[-_ ]', '-', regex=True)
crawled_data['model'].value_counts().head(50)
```

```python
# Extract horsepower from patterns like "60 kW (82 hp)"
crawled_data['power'] = crawled_data['power'].astype(str).str.extract(r'\((\d+)\s*hp\)').astype(float)
crawled_data['power'].describe()
```

```python
kaggle_data.rename(columns={'powerPS': 'power'}, inplace=True)
kaggle_data['power'].describe()
```

```python
target_cols = ['brand', 'model', 'vehicleType', 'power', 'yearOfRegistration', 'kilometer', 'fuelType', 'seller', 'price']
df_kaggle_final = kaggle_data[target_cols].copy()
df_crawled_final = crawled_data[target_cols].copy()
```

```python
df_merged = pd.concat([df_kaggle_final, df_crawled_final], ignore_index=True)
print(f"Shape of the dataset: {df_merged.shape}")
print(df_merged.info())
```

```python
os.makedirs('../data/clean', exist_ok=True)
df_merged.to_csv('../data/clean/merged_data.csv', index=False)
```
